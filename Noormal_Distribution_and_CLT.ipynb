### Central Limit Theorem:
- The mean of sample means is equal to population mean.
- The histogtam/distribution of sample means is looks like Normal Distribution.
(if sample size is high)

```python
df.sample(1000,random_state=1)
```

```python
df[df['cut'] == 'Good'].sample(1000)
```

```python
pop_mean = df['price'].mean()
pop_mean
```

```python
sample_means = []
for i in range(1000):
    sample_means.append(df['price'].sample(1000).mean())
```

```python
np.array(sample_means).mean()
```

```python
sns.histplot(sample_means,kde = True)
```

```python
def central_limit_theorem(data,col,N,n):
    pop_mean = data[col].mean()
    print('mean of {} is {}'.format(col,pop_mean))
    sample_means = []
    for i in range(N):
        sample_means.append(data[col].sample(n).mean())
    print('mean of sample means is {}'.format(np.array(sample_means).mean()))
    sns.histplot(sample_means,kde = True)
    plt.show()
```

```python
central_limit_theorem(df,'price',1000,1000)
```

 ### Emperical Rule 
 - The area in between $  \mu \pm \sigma$ is 68.45%
 - The area in between $  \mu \pm 2\sigma $ is 95% 
 - The area in between $  \mu \pm 3\sigma  $ is 99.73%
