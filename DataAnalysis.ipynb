### Outliers in Duration Colummn
#### Inter Quartile Range Method

iqr = q3 - q1

lb = q1 - 1.5*iqr

ub = q3 + 1.5*iqr

```python
q1 = df['duration'].quantile(0.25)
q3 = df['duration'].quantile(0.75)
iqr = q3 - q1

lb = q1 - (1.5*iqr)

ub = q3 + (1.5*iqr)
outliers = df[(df['duration']<lb) | (df['duration']>ub)]

q1,q3, iqr, lb, ub,len(outliers)
```

```python
def iqr_method(data,col):
    q1 = data[col].quantile(0.25)
    q3 = data[col].quantile(0.75)
    iqr = q3 - q1

    lb = q1 - (1.5*iqr)
    ub = q3 + (1.5*iqr)
    outliers = data[(data[col]<lb) | (data[col]>ub)]
    print('coulmn name:',col)
    print('number of the outliers:',len(outliers))
    print('% of the outliers:',len(outliers)/len(data))    
```

```python
iqr_method(df,'duration')
```

```python
iqr_method(df,'salary')
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X = X.drop('z',axis = 1)
```

```python
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
```
